## 1. Definição do Problema

O dataset utilizado no escopo deste projeto foi o MICRODADOS_ENEM_2021, disponibilizado no site https://www.gov.br/inep/pt-br/acesso-a-informacao/dados-abertos/microdados/enem pelo Instituto Nacional de Estudos e Pesquisas Educacionais Anisio Teixeira (INEP) O Exame Nacional do Ensino Médio (Enem) tem como objetivo primordial aferir se aqueles que dele participam demonstram, ao final do ensino médio, individualmente, domínio dos princípios científicos e tecnológicos que presidem a produção moderna e se detêm conhecimento das formas contemporâneas de linguagem. Os microdados do Enem são o menor nível de desagregação de dados recolhidos por meio do exame. O Data set apresenta muitos atributos relacionados a faixa etaria, sexo, cor/raça , tipo de escola que estudou, ano de conclusão do ensino médio, a pontuação obtida nas áreas de conhecimento e redação. O Problema apresentado é de aprendizado não supervisionado com a caracteristica de clusterização. Dependendo da forma de cálculo de pontuação por universidade, o candidato é classificado ou não. Queremos nesse estudo fazer correlaçãoes entre a pontuação entre áreas, por estado/municipio. A relação de notas entre o questionário socio-economico e a pontuação nas áreas de conecimento e redação. Os dados são anonimizaveis, não sendo possivel identificar um candidato ou mesmo a Instituição que o aluno estudou.

## 4. Pré-Processamento de dados



### 4.1. Tratamento de Missings e Limpeza

O datset do microdados do enem existem muitos dados nulos de candidatos que faltaram as provas ou foram eliminados.

```
# verificando nulls no dataset
enem_br[notas_cols].isnull().sum()
```

Existem quase 1/3 das notas nulas. Vamos criar um novo dataset  com quem estava não faltou a prova e tem nota valida.

```
# Selecione as linhas que não possuem notas nulas e que não têm faltas registradas
enem_br_sem_nulos = enem_br.loc[(enem_br['NU_NOTA_CN'].notnull()) & 
                                (enem_br['NU_NOTA_CH'].notnull()) & 
                                (enem_br['NU_NOTA_LC'].notnull()) & 
                                (enem_br['NU_NOTA_MT'].notnull()) &
                                (enem_br['TP_PRESENCA_CN'] == 1) & 
                                (enem_br['TP_PRESENCA_CH'] == 1) & 
                                (enem_br['TP_PRESENCA_LC'] == 1) &
                                (enem_br['TP_PRESENCA_MT'] == 1) &
                                (enem_br['faltou'] == False)]

# verificando nulls no dataset
enem_br_sem_nulos[notas_cols].isnull().sum()
```

Verificando o tamanho do novo dataset

```
enem_br_sem_nulos.shape
```

Vamos mostrar o dataset original em uma visualização matricial.


```
# salvando um NOVO dataset para tratamento de missings (cuidado para não sobrescrever o dataset original!)

# recuperando os nomes das colunas
#col = list(enem_br[notas_cols])

# o novo dataset irá conter todas as colunas referente a notas
atributos = enem_br[notas_cols]

# exibindo visualização matricial da nulidade do dataset
ms.matrix(atributos)
```
